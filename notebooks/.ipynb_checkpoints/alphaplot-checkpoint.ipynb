```python
import warnings
warnings.filterwarnings("ignore")
%load_ext autoreload
%autoreload 2
%matplotlib inline
import numpy as np
import h5py
import plotly
import matplotlib.pyplot as plt
import plotly.plotly as py
import yaml
import sys
sys.path.append('..')
from src.models.CRsAE import *
PATH = sys.path[-1]
```

```python
# load data
print("load data.")
folder_name = "alphaplot_alpha1.00_filter3_fdim48_idim3000_train10000_test100_cycleLR"
hf = h5py.File("{}/experiments/{}/data/data.h5".format(PATH, folder_name), "r")
y_train = np.array(hf.get("y_train"))
y_test = np.array(hf.get("y_test"))
z_train = np.array(hf.get("z_train"))
z_test = np.array(hf.get("z_test"))
y_train_noisy = np.array(hf.get("y_train_noisy"))
y_test_noisy = np.array(hf.get("y_test_noisy"))
noiseSTD = np.array(hf.get("noiseSTD"))

H_true = np.load("{}/experiments/{}/data/H_true.npy".format(PATH, folder_name))

plt.plot(np.squeeze(H_true))
```

```python
# load model parameters
print("load model parameters.")
file = open("{}/experiments/{}/config/config_model.yml".format(PATH, folder_name), "rb")
config_m = yaml.load(file)
file.close()

# load data parameters
print("load data parameters.")
file = open("{}/experiments/{}/config/config_data.yml".format(PATH, folder_name), "rb")
config_d = yaml.load(file)
file.close()

# create CRsAE object
print("create CRsAE object.")
if config_m["data_space"] == 1:
    crsae = CRsAE_1d(
        config_m["input_dim"],
        config_m["num_conv"],
        config_m["dictionary_dim"],
        config_m["num_iterations"],
        config_m["L"],
        config_m["twosided"],
        config_m["MIMO"],
        config_m["alpha"],
        config_m["num_channels"],
    )
    
# build model knowing noiseSTD
crsae.build_model(noiseSTD)

# set H
crsae.set_H(H_true)

#predict
i = 48
print("do prediciton.")
z_test_hat = crsae.encode(np.expand_dims(y_test_noisy[i,:,:],axis=0))
y_test_hat = crsae.denoise(np.expand_dims(y_test_noisy[i,:,:],axis=0))

fig = plt.figure(figsize=(20,5))
plt.plot(y_test_noisy[i,:,0],color='g',label='$y_{noisy}$')
plt.plot(y_test[i,:,0],color='k',label='$y$')
plt.plot(np.squeeze(y_test_hat),color='r',label='$\hat y$')
plt.legend()

fig = plt.figure(figsize=(20,5))
for n in range(config_m["num_conv"]):
    plt.subplot(1,config_m["num_conv"],n+1)
    plt.plot(z_test[i,:,n],'ok',label='$z$')
    plt.plot(z_test_hat[0,:,n],'.r',label='$\hat z$')
    plt.plot(z_test_hat_old[0,:,n],'.g',label='$\hat z$')
plt.legend()


z_test_hat_old = z_test_hat
```
